### 데모 - Slack Scheduler bot 를 만들 예정입니다- 다른 구글 캘린더 사용자와의 미팅과 리마인더 스케줄링을 위한 언어 인식 지능 slack bot 입니다. 다른 매체 (이메일, 데이터베이스, whatsapp 등) 에도 쉽게 적용 가능합니다. 

## 구인 관련 문제 

![alt text](https://ideal.com/wp-content/uploads/2016/11/recruiting-AI-technology.png "Logo Title Text 1")

- 인터뷰 스케줄링은 시간을 많이 소요하는 작업입니다. 이상적으로 자동화할 수 있으며 시간과 돈을 저축할 수 있고 결과적으로 생산성을 높일 수 있습니다. 
- 구인 산업에서 생산성을 어떻게 높일 수 있을까요?
- 우리는 사용 가능한 수많은 AI 도구를 가지고 있으며, 이러한 문제 해결을 위한 해결책을 자체 개발할 수 있습니다. 
- Slack Scheduler Bot 은 크고 immediately addressable한 시장 (미팅 일정을 잡고 Slack을 사용하는 팀) 의 업무량을 정확한 시간 내에 간소화함으로써 실제 사용자 문제를 해결합니다.
- 구인 팀에 쉽게 적용 가능합니다.

## 구인의 역사 

![alt text](https://img.etimg.com/photo/60986101/pg-4-1.jpg "Logo Title Text 1")
![alt text](https://img.etimg.com/photo/60986105/pg-4-2.jpg "Logo Title Text 1")

-  HR 팀과 구인 담당자는 종이, 팩스, 이메일 이력서의 시대를 거쳐 왔습니다. 
-  90 년대 이전: 종이 이력서, 수동 데이터 업로드, 비효율적이고 값비싼 직업 박람회. 
-  1999 년: Monster.com 과 Indeed.com: 최초 인터넷 구인 사이트. 사람들이 일자리를 찾기 쉽게 해주었으나, 구인하는 회사들에는 별다른 이점을 제공하지 않았습니다. 
-  2000 년대 초: 사람들은 회사 웹사이트에서 구직하였고, 몇몇은 회사 웹 포털에서 구직하였으나, 원하는 자료를 찾기 힘들었습니다. 
- 2008 년: 금융의 하락세가 구인의 변화를 이끌었습니다. 회사는 지원자를 평가하기 위해 더 까다롭게 선택해야 했습니다. 
- 2010 년 - 현재: 지원자에 대한 정보가 많습니다. AI가 구인 과정의 첫 파트에서 통찰력을 제공하는 데 사용될 수 있습니다. 
- AI는 구인 과정의 두 번째 파트에서 적절한 후보자와의 스케줄을 잡는 데 사용될 수 있습니다.

#### 구인 산업에 영향을 주는 중요한 요소 

- 베이비 붐 세대들이 은퇴하기 시작하였으며 일자리에 공석이 생기고 있습니다.
- SaaS 가 다시 떠오르며, 두 번째 기술 붐을 일으키고 있습니다. 
- 소셜 미디어가 흔하게 쓰이고 있습니다.
- 젊은 세대들은 구직할 때 신문과 지면 광고를 이용하지 않고, 온라인 직업 게시판, 소셜 미디어, 그리고 회사 내 구인 웹사이트를 이용합니다.

##### 직업 박람회와 대면 이력서 drop-off는 과거의 일이지만, 이력서 업로드의 크기는 증가하고 있습니다. 매달 수백 수천 개의 구직자 중 완벽한 지원자, 혹은 인터뷰가 준비된 지원자를 찾는 것조차도 매우 힘듭니다.

- 구인 소프트웨어는 지능적인 검색과 필터링 함수를 통해 구인자가 실제로 읽을 만한 이력서의 수를 줄여줍니다. 
- 즉, 학위, 키워드, 온라인 이력서 데이터베이스 경험의 수준 등을 검색하여 더 나은 후보자를 빠르게 찾습니다.

## 아키텍쳐

###### 우리의 인공지능 bot은 다음 기술을 사용합니다 

![alt text](https://files.readme.io/a769bab-API-AI_key-concepts.png "Logo Title Text 1")

![alt text](http://www.3bluefrogs.com/wp-content/uploads/2016/01/1452909989test1-1024x453.jpg "Logo Title Text 1")

- API.AI - 자연어 처리를 위한 인공지능/기계학습 도구 
- Slack - RTM과 상호 메세지를 위한 대화 앱 
- Google Calendar API - 미팅과 리마인더 스케줄링에 bot이 사용함 
- Express - 백엔드 서버

##### 단계

1. Slack bot 이 다음과 같은 메세지를 보내면 "특정 사람과 특정 시간에 특정 날짜에 미팅 주제를 얘기 나누기 위해 1 시간 동안 미팅 약속을 잡아주세요" 이를 즉시 API.AI 로 보내 이를 분석하여 미리 저장된 의도를 파악한 뒤 사용자의 입력을 기반으로 반응을 어떻게 할 지 결정합니다 
2. 그 다음 Scheduler bot 은 양 사용자의 캘린더에 일정을 추가하기 위한 권한이 있는지 확인합니다. 권한이 없다면, 요청합니다.
3. 그런 다음 양 사용자의 캘린더를 확인하여 충돌이 있는지 확인합니다. 선약이 있다면, bot은 다시 사용자에게 충돌이 있으니 슬롯 리스트에서 새 시간을 선택하라고 답장합니다.
4. 사용자는 최적의 시간을 선택하고 bot 은 미팅을 잡고 양 사용자의 캘린더에 일정을 추가한 뒤 이메일로 알립니다.

- 스케줄링과 같은 방식으로 리마인더도 생성 가능하며, bot은 사용자의 캘린더에 리마인더를 설정합니다.
- 이는 Heroku 의 스크립트 스케줄러 script.js 파일을 통해 가능하며, 매일 밤 자정에 cron 작업을 수행합니다.
